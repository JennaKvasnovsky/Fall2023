```python
best_loss = 1000000

for m in M:
    for b in B:
        m_guess = m
        b_guess = b
        y_guess = m_guess * x + b_guess

        loss = np.sum((y - y_guess)**2)

        if loss < best_loss:
            best_loss = loss
            print(m,b,loss)
```
